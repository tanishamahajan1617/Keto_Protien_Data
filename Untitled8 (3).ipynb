```python
import requests as rq
import pandas as pd
import numpy as np


url = 'https://raw.githubusercontent.com/tanishamahajan1617/Keto_Protien_Data/main/_MConverter.eu_jsonformatter.json'


headers = {
    'Accept': 'application/json',
    'User-Agent': 'Mozilla/5.0'
}


response = rq.get(url,headers = headers)


```

```python
data = response.json()
df = pd.DataFrame(data)
df
```

```python

df.to_csv(r'C:\Projects\dataset\keto_recipes.csv', index=False)
```
